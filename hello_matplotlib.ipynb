Import NumPy and matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

Set Matplotlib to see the result in this window

```python
%matplotlib inline
```

Create some sample data

```python
# On X-axis number from 1 to 15
a = range(1,16)

# On Y-axis generate random integer with mean and sigma (standard deviation)
mean = 50
sigma = 10

b = np.random.normal(mean, sigma, 15).astype(int)
```

Plot the chart

```python
plt.plot(a,b)
```

Change the color of lines

```python
# Define Color
plt.plot(a,b,color='red')
```
